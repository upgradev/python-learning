# Generate a normal random distribution of order 20 x 5, with a 
# mean of 10 and standard deviation 3. Store the result in a variable. Then create a DataFrame from the data created.



```python
import numpy as np
import pandas as pd
```

```python
mean = 10
standard_deviation = 3
data = np.random.normal(loc=mean, scale=standard_deviation, size=(20,5))
df = pd.DataFrame(data, columns=list('ABCDE'), index=range(1, data.shape[0]+1))
df
```
