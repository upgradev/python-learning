```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the dataset
data=pd.read_csv("bigmart.csv")
```

```python
data.head()
```

```python
#checking for univariate outliers
plt.rcParams["figure.figsize"]=(16,4)

plt.subplot(1,4,1)
sns.boxplot(x = data["Item_Weight"], orient="h")

plt.subplot(1,4,2)
sns.boxplot(x = data["Item_Visibility"], orient="h")

plt.subplot(1,4,3)
sns.boxplot(x = data["Item_MRP"], orient="h")

plt.subplot(1,4,4)
sns.boxplot(x = data["Item_Outlet_Sales"], orient="h")

plt.suptitle("Checking for Univariate Outliers")
plt.show()
```

```python
plt.subplot(1,3,1)
sns.scatterplot(y=data["Item_Outlet_Sales"], x=data["Item_Weight"])

plt.subplot(1,3,2)
sns.scatterplot(y=data["Item_Outlet_Sales"], x=data["Item_Visibility"])

plt.subplot(1,3,3)
sns.scatterplot(y=data["Item_Outlet_Sales"], x=data["Item_MRP"])

plt.suptitle("Checking for Univariate Outliers")
plt.show()
```

# Capping outliers

capping refers to replacing the outliers to a near values so that we can keep the point in our analysis and it also does not skew the data

```python
#capping the putliers in the column Item_Outlet_Sales having values greater than 10000
data["Item_Outlet_Sales"].values[data["Item_Outlet_Sales"].values > 10000]=10000
```

```python
#Checking if there is still any greater than 10000 left in the Item_Outlet_Sales column
data[data["Item_Outlet_Sales"]>10000]
```

```python
#dealing with the bivariate outliers
#we delete all the values which are greater than 8000
data = data[~(data["Item_Outlet_Sales"]>8000)]
```
